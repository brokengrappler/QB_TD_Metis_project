```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
sns.set()
```

```python
path = './Raw_Data/players/'
```

```python
os.listdir(path)
```

```python
def player_df_dict():
    df_dict = {}
    for files in os.listdir(path):
        name = files.split('.')[0]
        df = pd.read_csv(path+files, skipfooter=1, engine='python')
        df_dict[name] = df
    return df_dict
```

```python
player_db = player_df_dict()
```
